```python
import pandas as pd
df = pd.read_csv("data/PEP_2017_PEPAGESEX.csv",header = 1)
df.columns.values
for i in range(len(df.columns.values)):
    print(i, df.columns.values[i])
```

```python
april_list = []
for i in range(len(df.columns.values)):
    if len(df.columns.values[i])>=5 and df.columns.values[i][:5] == 'April':
        april_list.append(df.columns.values[i])
```

```python
len(april_list)
```

```python
df_copy = df

```

```python
df_remove_april= df_copy.drop(columns = april_list)
```

```python
df_remove_april.columns
for i in range(len(df_remove_april.columns.values)):
    print(df_remove_april.columns.values[i])
    
```

```python
droplist2011=[]
droplist2012=[]
droplist2013=[]
droplist2015=[]
droplist2016=[]
len("Population Estimate (as of July 1) - 2010")
```

```python
for i in range(len(df_remove_april.columns.values)):
    if len(df_remove_april) >=5 and df_remove_april.columns.values[i][37:41]== '2011':
        droplist2011.append(df_remove_april.columns.values[i])

```

```python
for i in range(len(df_remove_april.columns.values)):
    if len(df_remove_april) >=5 and df_remove_april.columns.values[i][37:41]== '2012':
        droplist2012.append(df_remove_april.columns.values[i])
for i in range(len(df_remove_april.columns.values)):
    if len(df_remove_april) >=5 and df_remove_april.columns.values[i][37:41]== '2013':
        droplist2013.append(df_remove_april.columns.values[i])
for i in range(len(df_remove_april.columns.values)):
    if len(df_remove_april) >=5 and df_remove_april.columns.values[i][37:41]== '2015':
        droplist2015.append(df_remove_april.columns.values[i])
for i in range(len(df_remove_april.columns.values)):
    if len(df_remove_april) >=5 and df_remove_april.columns.values[i][37:41]== '2016':
        droplist2016.append(df_remove_april.columns.values[i])




```

```python
df_new = df_remove_april.drop(columns = droplist2011)
df_new1 = df_new.drop(columns= droplist2012)
df_new2 = df_new1.drop(columns = droplist2013)
df_new3 = df_new2.drop(columns = droplist2015)
df_new4 = df_new3.drop(columns = droplist2016)

df_new5 = df_new4.drop(columns = ["Id",'Id2','Geography'])
df_new5.shape
```
